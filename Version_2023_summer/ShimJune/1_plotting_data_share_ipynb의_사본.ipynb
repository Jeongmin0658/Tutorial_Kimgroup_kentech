# Part I. Plotting data

# 1 - Below is an example to plot y=x^2.

```python
# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 100)

# Compute the corresponding y values
y = x ** 2

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels for x and y axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a title
ax.set_title('$y = x^2$')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 20000)

y1 = np.exp(-x**2)
y2 = np.exp(-x**2/5)
y3 = np.exp(-x**2/10)


fig, ax = plt.subplots()
line1, = ax.plot(x,y1)
line2, = ax.plot(x,y2)
line3, = ax.plot(x,y3)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('$y=exp(-x^2/t)$')

```
