# Part I. Plotting data

# 1 - Below is an example to plot y=x^2.

```python
#upload test

# import necessary libraries
import matplotlib.pyplot as plt #plot
import numpy as np #number

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 100)

# Compute the corresponding y values
y = x ** 2

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels for x and y axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a title
ax.set_title('$y = x^2$')

# Display the plot
plt.show()
```

(a) $y=x^2+4x5$ (-5<x<10)

    (Optional) Plot the solutions of this quadatic equation as points.



```python
# import necessary libraries
import matplotlib.pyplot as plt #plot
import numpy as np #number

# Create an array of x values from -5 to 10
x = np.linspace(-5, 10, 100)

# Compute the corresponding y values
y = x**2 + 4*x - 5

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels for x and y axes
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

# Add a title
ax.set_title('$=^2+45$')

# Display the plot
plt.show()
```

(b) $y=7*\cos(\pi x)$ (0<x<10)

```python
# import necessary libraries
import matplotlib.pyplot as plt #plot
import numpy as np #number

# Create an array of x values from 0 to 10
x = np.linspace(0, 10, 100)

# Compute the corresponding y values
y=7*np.cos(np.pi*)

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels for x and y axes
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

# Add a title
ax.set_title('$=7cos()$')

# Display the plot
plt.show()
```

    (d-i)(0<x<100)

```python
#d-i

# import necessary libraries
import matplotlib.pyplot as plt #plot
import numpy as np #number

# Create an array of x values from 0 to 10
x = np.linspace(0, 100, 100)

# Compute the corresponding y values
=-np.log10()

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels for x and y axes
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

# Add a title
ax.set_title('$=log10()$')

# Display the plot
plt.show()
```
