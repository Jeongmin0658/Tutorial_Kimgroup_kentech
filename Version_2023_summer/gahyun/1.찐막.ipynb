# Part I. Plotting data

# 1 - Below is an example to plot y=x^2.

```python
# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

#aklsjfsald

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 100)

# Compute the corresponding y values
y = x ** 2

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels for x and y axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a title
ax.set_title('$y = x^2$')

# Display the plot
plt.show()
```

```python
# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

#aklsjfsald

# Create an array of x values from -10 to 10
x = np.linspace(-5, 10, 100)

# Compute the corresponding y values
y = x ** 2 + 4*x -5

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)
ax.plot(-5,0,"o")
ax.plot(1,0,"o")

# Add labels for x and y axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a title
ax.set_title('$y = x^2 +4x -5$')

# Display the plot
plt.show()
```

```python
# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

#aklsjfsald

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 100)
k=x**2

# Compute the corresponding y values
y = np.exp(-k)
y1 = np.exp(-k/5)
y2 = np.exp(-k/10)

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()
D=["D=1", "D=5", "D=10"]

# Plot the data
ax.plot(x, y)
ax.plot(x, y1)
ax.plot(x, y2)
plt.yscale('log')
# Add labels for x and y axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a title
ax.set_title('$y = exp(-x^2/D)$')

# Display the plot
plt.legend(D)
plt.show()
```
