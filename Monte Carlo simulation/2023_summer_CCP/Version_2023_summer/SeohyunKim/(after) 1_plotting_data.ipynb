# Part I. Plotting data

# 1 - Below is an example to plot y=x^2.

```python
# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 100)

# Compute the corresponding y values
y = x ** 2

# Create a figure and a set of subplots (this is often combined in one step)
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels for x and y axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a title
ax.set_title('$y = x^2$')

# Display the plot
plt.show()
```

# 3 - You have more pratice examples.

(a) $y=x^2+4x5$ (-5<x<10)

    (Optional) Plot the solutions of this quadatic equation as points.


```python
import math
```

```python
def sol(a,b,c):
  x1 = (-b + math.sqrt(b**2 - 4*a*c)) / 2*a
  x2 = (-b - math.sqrt(b**2 - 4*a*c)) / 2*a
  return np.array([x1,x2])

x = np.linspace(-5,10,100)
y = x**2 + 4*x -5

sol_x = sol(1,4,-5)
sol_y = sol_x**2 + 4*sol_x -5

temp_x = np.linspace(-10,-5,100)
temp_y = temp_x**2 + 4*temp_x -5

fig, ax = plt.subplots()

ax.plot(x, y)
ax.scatter(sol_x,sol_y,color = 'r')
ax.plot(temp_x,temp_y,'--')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('$y = x^2 + 4x - 5$')
plt.show()
```

(b) $y=7*\cos(\pi x)$ (0<x<10)

```python
x = np.linspace(0,10,100)
y = 7 * np.cos(np.pi * x)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('$y = 7cos(pi * x)$')
plt.show()
```

(d) $y=-\log_{10}(x)$

    (d-i)(0<x<100)
    
    (d-ii)(0<x<1000 and set x axis to logscale)

```python
x = np.linspace(0.1,100,100)
y = - np.log10(x)

fig, ax = plt.subplots(2,1)

ax[0].plot(x, y)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('$y = -log10(x)$')

ax[1].plot(x,y)
ax[1].set_xscale('log')
ax[1].set_xlabel('x (log scale)')
ax[1].set_ylabel('y')
ax[1].set_title('$y = -log10(x)$')

plt.tight_layout()
plt.show()
```

(e) $y = \exp(-x/t)$ with $t$ = 1,10,100

    (e-i) Linear x axis, Linear y axis
    
    (e-ii) Log x axis, Linear y axis

```python
x = np.linspace(0,100,100)
y1 = np.exp(-x / 1)
y2 = np.exp(-x / 10)
y3 = np.exp(-x / 100)

_x = np.linspace(0,100,10)
_y1 = np.exp(-_x / 1)
_y2 = np.exp(-_x / 10)
_y3 = np.exp(-_x / 100)

fig, ax = plt.subplots(2,1)

line1, = ax[0].plot(x, y1, label = 't=1')
ax[0].scatter(_x, _y1, label = 't=1')
line2, = ax[0].plot(x, y2, label = 't=10')
ax[0].scatter(_x, _y2, label = 't=10')
line3, = ax[0].plot(x, y3, label = 't=100')
ax[0].scatter(_x, _y3, label = 't=100')
ax[0].legend()
ax[0].set_xlabel('x (linear scale)')
ax[0].set_ylabel('y (linear scale)')
ax[0].set_title('$y = exp(-x/t)$')
ax[0].legend(handles=[line1, line2, line3])

line1, = ax[1].plot(x, y1, label = 't=1')
ax[1].scatter(_x, _y1, label = 't=1')
line2, = ax[1].plot(x, y2, label = 't=10')
ax[1].scatter(_x, _y2, label = 't=10')
line3, = ax[1].plot(x, y3, label = 't=100')
ax[1].scatter(_x, _y3, label = 't=100')
ax[1].legend()
ax[1].set_xscale('log')
ax[1].set_xlabel('x (log scale)')
ax[1].set_ylabel('y (linear scale)')
ax[1].set_title('$y = exp(-x/t)$')
ax[1].legend(handles=[line1, line2, line3])

plt.tight_layout()
plt.show()
```

(f) $y = \exp(-x^2/D)$ with $D$ = 1,5,10 ( -10 < x < 10 )

    (Optional) Think about how to present data effectively

```python
def func(x,D):
  return np.exp(- x**2 / D)

x = np.linspace(-10,10,100)
y1 = func(x,1)
y2 = func(x,5)
y3 = func(x,10)

fig, ax = plt.subplots(3,1,figsize=(7,10))

ax[0].plot(x, y1, label = 'D=1')
ax[0].plot(x, y2, label = 'D=5')
ax[0].plot(x, y3, label = 'D=10')
ax[0].legend()
ax[0].set_xlabel('x (linear scale)')
ax[0].set_ylabel('y (linear scale)')
ax[0].set_title('$y = exp(-x**2/D)$')

ax[1].plot(x, y1, label = 'D=1')
ax[1].plot(x, y2, label = 'D=5')
ax[1].plot(x, y3, label = 'D=10')
ax[1].legend()
ax[1].set_xscale('log')
ax[1].set_xlabel('x (log scale)')
ax[1].set_ylabel('y (linear scale)')
ax[1].set_title('$y = exp(-x**2/D)$')

ax[2].plot(x, y1, label = 'D=1')
ax[2].plot(x, y2, label = 'D=5')
ax[2].plot(x, y3, label = 'D=10')
ax[2].legend()
ax[2].set_yscale('log')
ax[2].set_xlabel('x (linear scale)')
ax[2].set_ylabel('y (log scale)')
ax[2].set_title('$y = exp(-x**2/D)$')

plt.tight_layout()
plt.show()
```
